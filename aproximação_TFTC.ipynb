# Aproximação de uma Transformadda de Fourier de Tempo Contínuo (TFTC)
Nesse código iremos amostrar uma função de tempo contínuo e depois aproximar a TFTC. A amostragem da função de tempo contínuo irá gerar uma sequência de duração finita.

##Importações:


```
import numpy as np
import matplotlib.pyplot as plt
# ...
```

## Definição de sequências n (índices de tempo discreto), k (índices de frequências discretizadas) e dos valores de omega_k:
Os índices de tempo vão de -0.005 a 0.005 (com delta_t igual a 0.00001, isto é,  0.005/500), enquanto que os índices de frequência vão de -M a M-1, onde M é arbitrário. Os valores de omega_k é igual a pi multiplicado por k, e dividido por M.

```
K = 1000
m_delta_t = np.arange(-0.005,0.005,0.00001).reshape(1,1000) #1000 elementos
k = np.array(range(-K,K)).reshape(1,2*K)
omega_k = np.array(4*np.pi*(k/K))
```

## Definição da sequência x_v[m_delta_t]: 

```
x_v = np.exp(-1000*np.abs(m_delta_t))
x_v.shape
```

##Definição da rotina que retorna a transformada de Fourier:

```
def tctf(x,n,w):
  x_f = np.empty((1,np.shape(w)[1])) # a quantidade de linhas da transformada é igual a quantidade de linhas do vetor omega
  matriz_fourier = np.exp(-(0+1j)*n.T*w)
  x_f = np.dot(x,matriz_fourier)
  return x_f

  
```

##Funções de Plotagem:

```


def plot_mag(w,x_fourier):
  plt.title("Magnitude x omega")
  plt.xlabel("frequency [rad]")
  plt.ylabel("magnitude")
  plt.plot(w[0,:],np.abs(x_fourier)[0,:])
  plt.show()
  
def plot_freq(w,x_fourier):
  plt.title("Phase x omega")
  plt.xlabel("frequency [rad]")
  plt.ylabel("phase")
  plt.plot(w[0,:],np.angle(x_fourier)[0,:])
  plt.show()
```

##TESHTANDO

```
fou = tctf(x_v,m_delta_t,omega_k)
plot_freq(omega_k,fou)
plot_mag(omega_k,fou)
```
